```python
from Bio.Seq import Seq
```

```python
DNA = Seq("ATGATCTCGTAA")
print (f"DNA: {DNA}")

RNA = DNA.transcribe()
print(f"RNA: {RNA}")

Protein = RNA.translate()
print(f"Protein: {Protein}")
```

```python
print(f"DNA Slicing: {DNA[0:3]}")
print(f"DNA Reverse: {DNA[::-1]}")
print(f"Panjang DNA: {len(DNA)}")
```

```python
print(f"Banyak G dalam {DNA}: {DNA.count('G')}")
print(f"Banyak T dalam {DNA}: {DNA.count('T')}")
print(f"Index G dalam {DNA}: {DNA.index('G')}")
print(f"Index T dalam {DNA}: {DNA.index('T')}")
print(f"Indeks ke-3 dalam {DNA}: {DNA[3]}")
print(f"G dari kanan: {DNA.rindex('G')}")
print(f"Mencari 'Z' (tidak ada): {DNA.find('Z')}")
print(f"rfind('G'): {DNA.rfind('G')}")
print(f"\nATG ditemukan pada index: {DNA.find('ATG')}")
```

```python
DNA1 = Seq("ATGCATGGTGGCGGA")
DNA2 = Seq("ATTTGTGCTCCTGGA")

print("DNA1:", DNA1)
print("DNA2:", DNA2)
```

```python
def nucleotide_frequency(seq):
  return {
"A": seq.count("A"),
"T": seq.count("T"),
"G": seq.count("G"),
"C": seq.count("C")
}

freq_DNA1 = nucleotide_frequency(DNA1)
freq_DNA2 = nucleotide_frequency(DNA2)

print("Frequency DNA1:", freq_DNA1)
print("Frequency DNA2:", freq_DNA2)
```

```python
plt.bar(freq_DNA1.keys(), freq_DNA1.values())
plt.title("Nucleotide Frequency - DNA1")
plt.xlabel("Nucleotide")
plt.ylabel("Count")
plt.show()
```

```python
plt.bar(freq_DNA2.keys(), freq_DNA2.values())
plt.title("Nucleotide Frequency - DNA2")
plt.xlabel("Nucleotide")
plt.ylabel("Count")
plt.show()
```

```python
from Bio.SeqUtils import gc_fraction
from Bio.SeqUtils import MeltingTemp, molecular_weight
```

```python
gc_DNA1 = gc_fraction(DNA1) * 100
gc_DNA2 = gc_fraction(DNA2) * 100
print("GC Content DNA1:", gc_DNA1, "%")
print("GC Content DNA2:", gc_DNA2, "%")
```

```python
tm_DNA1 = MeltingTemp.Tm_Wallace(DNA1)
tm_DNA2 = MeltingTemp.Tm_Wallace(DNA2)
print("Melting Temperature DNA1:", tm_DNA1, "°C")
print("Melting Temperature DNA2:", tm_DNA2, "°C")
```

```python
mw_DNA1 = molecular_weight(DNA1, "DNA")
mw_DNA2 = molecular_weight(DNA2, "DNA")
print("Molecular Weight DNA1:", mw_DNA1)
print("Molecular Weight DNA2:", mw_DNA2)
```

```python
print("KESIMPULAN:")
if gc_DNA1 > gc_DNA2:
  print("GC Content tertinggi: DNA1")
else:
  print("GC Content tertinggi: DNA2")
if tm_DNA1 > tm_DNA2:
    print("Melting Temperature tertinggi: DNA1")
else:
    print("Melting Temperature tertinggi: DNA2")
if mw_DNA1 > mw_DNA2:
    print("Molecular Weight tertinggi: DNA1")
else:
    print("Molecular Weight tertinggi: DNA2")
```
