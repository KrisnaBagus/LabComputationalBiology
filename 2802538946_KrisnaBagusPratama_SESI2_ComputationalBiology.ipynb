```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
```

```python
DNA = Seq( "ATGCGTAACG" )
print( f"DNA: {DNA}")
```

```python
def gc_count( seq ):
  guanine = seq.count( "G" )
  cytosine= seq.count( "C" )
  gc = ( guanine + cytosine ) / len( seq )
  return gc * 100
```

```python
def at_count( seq ):
  adenine = seq.count( "A" )
  thymine = seq.count( "T" )
  at = ( adenine + thymine ) / len( seq )
  return at * 100
```

```python
GC = gc_fraction( DNA )
print( f"GC dari {DNA}: {GC:.2f}")

GC = GC * 100
print( f"GC dari {DNA}: {GC:.0f}%")
```

```python
print( f"GC dari {DNA}: {gc_count(DNA):.0f}%")
print( f"AT dari {DNA}: {at_count(DNA):.0f}%")

print( f"GC + AT = {gc_count(DNA) + at_count(DNA)}%")
```

```python
def gc_loop( seq ):
  gc = []
  for basa in seq:
    if basa in "GC":
      gc.append(basa)
  return (len(gc) / len(seq)) * 100
```

```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
```

```python
DNA1 = Seq( "ATGCATGGTGCGCGA")
DNA2 = Seq( "ATTTGTGCTCCTGGA")
```

```python
print( f"DNA1: {DNA1}")
print( f"DNA2: {DNA2}\n")
```

```python
GC1 =gc_fraction( DNA1 )
GC2 =gc_fraction( DNA2 )
```

```python
print(f"GC dari DNA1: : {GC1*100:.2f}%")
print(f"GC dari DNA2: : {GC2*100:.2f}%\n")
```

```python
def gc_count(seq):
    guanine = seq.count("G")
    cytosine = seq.count("C")
    gc = (guanine + cytosine) / len(seq)
    return gc * 100

def at_count(seq):
    adenine = seq.count("A")
    thymine = seq.count("T")
    at = (adenine + thymine) / len(seq)
    return at * 100

print(f"GC dari (DNA1): {gc_count(DNA1):.0f}%")
print(f"AT dari (DNA1): {at_count(DNA1):.0f}%")
print(f"GC + AT DNA1 = {gc_count(DNA1) + at_count(DNA1)}%\n")

print(f"GC dari (DNA2): {gc_count(DNA2):.0f}%")
print(f"AT dari (DNA2): {at_count(DNA2):.0f}%")
print(f"GC + AT DNA2 = {gc_count(DNA2) + at_count(DNA2)}%\n")
```

```python
def gc_loop(seq):
    gc = []
    for basa in seq:
        if basa in "GC":
            gc.append(basa)
    return (len(gc) / len(seq) * 100)

print(f"GC Loop (DNA1): {gc_loop(DNA1):.0f}%")
print(f"GC Loop (DNA2): {gc_loop(DNA2):.0f}%\n")
```

## HASIL AKHIR

```python
print("=== HASIL AKHIR ===")

print("1. Which of the sequence will have the highest GC ? :\n",
        "DNA1" if gc_count(DNA1) > gc_count(DNA2) else "DNA2")

print("2. Which of the sequence will have higher melting temperature ? :\n",
        "DNA1" if MeltingTemp.Tm_Wallace(DNA1) > MeltingTemp.Tm_Wallace(DNA2) else "DNA2")

print("3. Which of the sequence wil have the highest Molecular Weight ? :\n",
        "DNA1" if MW1 > MW2 else "DNA2")
```
